## Loading and Visualizing Satellite Image

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show
import seaborn as sns

file_path = 'C:/Users/dell/Documents/GitHub/satellite-image-analysis/data/sample/sent2-4band_UKBMS_loc-0000_y-2019_m-06-09.tif'

with rasterio.open(file_path) as src:
    print(f"Image shape: {src.shape}")
    image = src.read()
    show(src, title="Original Satellite Image")



```

## Checking Bands and Metadata

```python
with rasterio.open(file_path) as src:
    print("Metadata:", src.meta)
    print("Number of bands:", src.count)
    print("CRS:", src.crs)
```

## Show Individual Bands

```python
fig, ax = plt.subplots(1, image.shape[0], figsize=(15, 5))
for i in range(image.shape[0]):
    ax[i].imshow(image[i], cmap='gray')
    ax[i].set_title(f'Band {i + 1}')
    ax[i].axis('off')
plt.tight_layout()
plt.show()
```
